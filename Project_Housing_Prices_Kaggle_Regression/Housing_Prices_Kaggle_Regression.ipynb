### Описание датасета:
* `SalePrice`: цена продажи недвижимости в долларах. Это целевая переменная, которую вы пытаетесь предсказать.
* `MSSubClass`: Класс здания
* `MSZoning`: Общая классификация зонирования
* `LotFrontage`: Линейные метры улицы, соединенной с недвижимостью
* `LotArea`: Размер участка в квадратных футах
* `Street`: Тип подъездной дороги
* `Alley`: Тип доступа к переулку
* `LotShape`: Общая форма собственности
* `LandContour`: Ровность участка
* `Utilities`: Тип доступных коммунальных услуг
* `LotConfig`: Конфигурация лота
* `LandSlope`: Уклон участка
* `Neighborhood`: Физическое местоположение в пределах города Эймс
* `Condition1`: Близость к главной дороге или железной дороге
* `Condition2`: Близость к главной дороге или железной дороге (при наличии второй)
* `BldgType`: Тип жилого помещения
* `HouseStyle`: Стиль жилища
* `OverallQual`: Общее качество материала и отделки
* `OverallCond`: Оценка общего состояния
* `YearBuilt`: Первоначальная дата постройки
* `YearRemodAdd`: Дата реконструкции
* `RoofStyle`: Тип крыши
* `RoofMatl`: Кровельный материал
* `Exterior1st`: Наружное покрытие дома
* `Exterior2nd`: Наружное покрытие дома (если используется более одного материала)
* `MasVnrType`: Тип облицовки каменной кладки
* `MasVnrArea`: Площадь облицовки каменной кладки в квадратных футах
* `ExterQual`: Качество наружного материала
* `ExterCond`: Текущее состояние материала на внешней поверхности
* `Foundation`: Тип фундамента
* `BsmtQual`: Высота подвала
* `BsmtCond`: Общее состояние подвала
* `BsmtExposure`: Стены подвала на уровне прохода или сада
* `BsmtFinType1`: Качество отделки подвала
* `BsmtFinSF1`: Готовые квадратные футы типа 1

* `BsmtFinSF2`: Готовые квадратные футы типа 2
* `BsmtUnfSF`: Незаконченные квадратные футы подвальной площади
* `TotalBsmtSF`: Общая площадь подвала в квадратных футах
* `Heating`: Тип отопления
* `HeatingQC`: Качество и состояние отопления
* `CentralAir`: Центральное кондиционирование воздуха
* `Electrical`: Электрическая система
* `1stFlrSF`: Площадь первого этажа в квадратных футах
* `2ndFlrSF`: Площадь второго этажа в квадратных футах
* `LowQualFinSF`: Низкое качество готовых квадратных футов (все этажи)
* `GrLivArea`: надземная жилая площадь в квадратных футах
* `BsmtFullBath`: Полностью оборудованные ванные комнаты на цокольном этаже
* `BsmtHalfBath`: Половина ванных комнат на цокольном этаже
* `FullBath`: Полноценные ванные комнаты высшего класса
* `HalfBath`: Половинные ванны выше класса
* `Bedroom`: Количество спален на цокольном этаже
* `Kitchen`: Количество кухонь
* `KitchenQual`: Качество кухни
* `TotRmsAbvGrd`: Общее количество номеров выше категории (не включает ванные комнаты)
* `Functional`: Рейтинг функциональности дома
* `Fireplaces`: Количество каминов
* `FireplaceQu`: Качество камина
* `GarageType`: Расположение гаража
* `GarageYrBlt`: Год постройки гаража
* `GarageFinish`: Внутренняя отделка гаража
* `GarageCars`: Размер гаража в зависимости от вместимости автомобиля
* `GarageArea`: Размер гаража в квадратных футах
* `GarageQual`: Качество гаража
* `GarageCond`: Состояние гаража
* `PavedDrive`: Мощеная подъездная дорожка
* `WoodDeckSF`: Площадь деревянной палубы в квадратных футах
* `OpenPorchSF`: Площадь открытой веранды в квадратных футах
* `EnclosedPorch`: Площадь закрытой веранды в квадратных футах
* `3SsnPorch`: Площадь веранды на три сезона в квадратных футах
* `ScreenPorch`: Площадь крыльца с экраном в квадратных футах
* `PoolArea`: Площадь бассейна в квадратных футах
* `PoolQC`: Качество бассейна
* `Fence`: Качество забора
* `MiscFeature`: Разные функции, не включенные в другие категории
* `MiscVal`: Значение разных функций
* `MoSold`: Месяц продажи
* `YrSold`: Год продажи
* `SaleType`: Тип продажи
* `SaleCondition`: Условие продажи

#### для baseline пройдемся по признакам в первом приближении т.к. их не так много.

```python
#сперва разделим все feutures на object и остальные
numeric_feutures = train_data.dtypes[train_data.dtypes != 'object'].index
object_feutures = train_data.dtypes[train_data.dtypes == 'object'].index
print(f'Получилось numeric_feutures {len(numeric_feutures)} шт. и object_feutures = {len(object_feutures)} шт.')
```

```python
#теперь проверим object_feutures на наличие пропусков
train_data[object_feutures].isna().sum()[train_data[object_feutures].isna().sum() > 0]
```

```python
#заменим пропущенные данные согласно описанию из data_description
train_data['Alley'] = train_data['Alley'].fillna('No alley access')
train_data['MasVnrType'] = train_data['MasVnrType'].fillna('None')
train_data[['BsmtQual','BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']] = train_data[['BsmtQual','BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']].fillna('No Basement')
train_data['FireplaceQu'] = train_data['FireplaceQu'].fillna('No Fireplace')
train_data[['GarageType','GarageFinish', 'GarageQual', 'GarageCond']] = train_data[['GarageType','GarageFinish', 'GarageQual', 'GarageCond']].fillna('No Garage')
train_data['PoolQC'] = train_data['PoolQC'].fillna('No Pool')
train_data['Fence'] = train_data['Fence'].fillna('No Fence')
train_data['MiscFeature'] = train_data['MiscFeature'].fillna('None')
```

После того как мы заменили пропуски согласно data_description, у нас остался только один неизвестный пропуск в 'Electrical'.

```python
#Теперь посмотрим на количество уникальных значений в object_feutures
unique_series = train_data[object_feutures].nunique().sort_values()

fig = plt.figure(figsize=(20, 4))
ax = sns.barplot(x=unique_series.index, y=unique_series.values, color='#5ef2c1')
ax.set_ylabel('Количество')
ax.set_xlabel('Признак')
ax.set_title('Количество уникальных значений в категориальных признаках')
plt.xticks(rotation = 90)

for i, val in enumerate(unique_series.values):
    plt.text(i, val, int(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':100, 'size':11})
```

Категориальных признаков достаточно много какие из них оставить а какие стоит переделать посмотрим чуть позже.

Теперь посмотрим на числовые признаки датасета

```python
#В первую очередь выделим из них категориальные признаки по описанию датасета
train_data[numeric_feutures].nunique()
```

```python
#список категориальных признаков из числовых
numeric_feutures_category =  ['MSSubClass', 'OverallQual', 'OverallCond', 'BsmtFullBath',
                              'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr',
                              'TotRmsAbvGrd', 'Fireplaces', 'GarageCars']

#список остальных фичей из числовых
numeric_feutures_other = [feutures for feutures in numeric_feutures if feutures not in numeric_feutures_category]
```

```python
#теперь посмотрим на пропуски в числовых признаках
train_data[numeric_feutures].isna().sum()[train_data[numeric_feutures].isna().sum() > 0]
```

```python
#заменим пропуски
#в MasVnrArea пропуски меняем на 0 если у того же объекта в MasVnrType стоит None
train_data['MasVnrArea'] = np.where((train_data['MasVnrArea'].isna()) & (train_data['MasVnrType']=='None'), 0, train_data['MasVnrArea'])

#в GarageYrBlt пропуски меняем на 0 если у того же объекта в GarageType стоит No Garage
train_data['GarageYrBlt'] = np.where((train_data['GarageYrBlt'].isna()) & (train_data['GarageType']=='No Garage'), 0, train_data['GarageYrBlt'])

```

Построим первую модель передсказаний на catboost

```python
from copy import deepcopy
data_for_model = deepcopy(train_data)
data_for_model['LotFrontage'] = data_for_model['LotFrontage'].median() #Много пропусков, поэтому пока что заполним медианным значением (изучим позже)
data_for_model = train_data.dropna(subset='Electrical') #Остался один пропуск который нужно удалить
```

```python
#Разбиваем на тестовую и валидационную выборку
#Стандартизацию не проводим т.к. будем использовать catboost - для него не требуется
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = data_for_model.drop(columns='SalePrice') #строю модель только на числовых фичах
y = data_for_model['SalePrice']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=2023)
```

# Baseline model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_log_error, mean_absolute_percentage_error
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split

cat_features = object_feutures.to_list() + numeric_feutures_category

#после небольшого иследования подобрали первые параметры для catboost
model_cat = CatBoostRegressor(verbose=False,
                              loss_function='MAE',
                              custom_metric=['RMSE', 'MAPE', 'MSLE'],
                              one_hot_max_size=9,
                              learning_rate=0.04,
                              depth=4
#                              one_hot_max_size=9)
)
model_cat.fit(X_train, y_train,
              cat_features=cat_features,
              eval_set=(X_valid, y_valid),
              plot=True)

predict_cat = model_cat.predict(X_valid)


print('MAE', '{:.4f}'.format(mean_absolute_error(y_pred=predict_cat, y_true=y_valid)))
print('RMSLE', '{:.4f}'.format(mean_squared_log_error(y_pred=predict_cat, y_true=y_valid, squared=False)))
print('MAPE', '{:.4f}'.format(mean_absolute_percentage_error(y_pred=predict_cat, y_true=y_valid)))
```

Отрисуем отдельно фичи у которых менее 6 признаков, более 5 признаков и отдельно числовые фичи которые мы определили как категориальные

```python
#Отрисовка категориальных фичей в которых менее 6 уникальных значений
fig = plt.figure(figsize=(15,30))

feutures_for_countplot_1 = unique_series[unique_series < 6].index

for index,col in enumerate(feutures_for_countplot_1):
    plt.subplot(11,5,index+1)
    fig.tight_layout(h_pad=1.2)
    ax = sns.countplot(data=train_data, x=col, palette='pastel')
    ax.set_ylabel(None)
    ax.set_xlabel(None)
    ax.set_title(col)
    plt.xticks(rotation = 90)
```

```python
#Блок для выделения фичей в которых преобладает одно значение
def clear_feutures(features):
    removed_feutures = []
    for i in features:
        count_in_feutures = train_data[i].value_counts().sort_values(ascending=False)
        big_feutures = count_in_feutures.iloc[0]
        if big_feutures / train_data.shape[0] * 100 > 85:
            removed_feutures.append(i)
    return removed_feutures
```

```python
removed_feutures_1 = clear_feutures(feutures_for_countplot_1)
len(removed_feutures_1)
```

```python
category_feutures = category_feutures + [feutures for feutures in feutures_for_countplot_1
                                         if feutures not in removed_feutures_1]
```

# Поиск оптимальных гиперпараметров

Будем искать оптимальные гиперпараметры с помощью встроенной в catboost функции кросс-валидации.

```python
# Определим вспомогательную функцию для поиска гиперпараметров по сетке
def grid_search_cv(grid, cv_dataset, iteration=1000, cv_fold=5, rs=2023):

    result = [['one_hot_max_size', 'lr', 'depth', 'RMSLE_score']]

    for lr in grid['learning_rate']:
        for depth in grid['depth']:
            for ohms in grid['one_hot_max_size']:
                params = {
                    'verbose': False,
                    'loss_function': 'MAE',
                    'custom_loss': 'MSLE',
                    'random_seed': rs,
                    'one_hot_max_size': ohms,
                    'learning_rate': lr,
                    'depth': depth,
                    'iterations': iteration
                }

                cv_data = cv(
                    params=params,
                    pool=cv_dataset,
                    fold_count=5,
                    partition_random_seed=rs)

                RMSLE_score = cv_data["test-MSLE-mean"].min()**0.5
                result.append([ohms, lr, depth, RMSLE_score])

    return result
```

```python
cv_dataset = Pool(data=X_cut,
                  label=y_cut,
                  cat_features=category_feutures)
```

```python
# Выполним поиск на сетке
grid = {'learning_rate': [0.03, 0.04],
        'depth': [3, 4],
        'one_hot_max_size': [7, 8, 9]}

grid_res = grid_search_cv(grid, cv_dataset=cv_dataset, iteration=4000,
                          cv_fold=5, rs=0)
```
